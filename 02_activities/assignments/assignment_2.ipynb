```python
# Given a list of numbers that represent a binary tree, with the first value representing the root, identify any duplicate(s) and return the value closest to the root. If no duplicate values exists, return -1.
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
# Input: [1,2,3,4,5]
# Output: [-1]

# The binary tree has a root node of 1. The next branches have nodes 2,3 - with child nodes 4,5 under 2. The tree has no duplicate values present, so -1 is returned.
```


-   Copy the solution your partner wrote. 


```python
def is_duplicate(root: TreeNode) -> int:  # O(n) x O(1) = O(n) algorithm
  # O(1) operations
  if not root:
     return -1
  seen = set()
  queue = [root]
  i = 0


  while i < len(queue):  # iterate over the nodes O(n)
     # O(1) operations
     node = queue[i]
     i += 1

     if node.val in seen:  # O(1) operation due to hash set
      return node.val
     
     seen.add(node.val)  # adding to hash set is O(1)
 
     # checking if node left/right exists is O(1) each
     if node.left:  
      queue.append(node.left)

     if node.right:
      queue.append(node.right)

  return -1

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(3)
root.right.right = TreeNode(2)

print(is_duplicate(root))
```


-   Explain why their solution works in your own words.


```python

# 1. Create a set to track values looped through and the root
# 2. Using a while loop, iterate through the tree values and track values 'seen'
# 3. If a value is found to be seen before, return it
# 4. If value hasn't been seen before, add to set and add children to the list/queue of numbers
# 5. If no duplicates found to be returned, then return -1
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time Complexity:
# The while loop - at worst case scenario (if no duplicates exist or the duplicate is found at the very end, visit each node exactly once) -  has O(n) time efficiency
# All other steps have an O(1) efficiency, and the while loop itself has O(n) time efficiency
# The queue list and seen set's size are each bounded by the number of nodes in the tree, so the algorithm has time efficiency O(n) 

# Space Complexity:
# The seen set can also grow up to n elements in the worst case - total space complexity is O(n)
```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


```python

# The solution uses BFS to traverse the tree level by level, keeping track of seen values in a hash set. 
# As soon as a value is encountered for the second time, the function returns that value. 
# If no duplicates are found after visiting all nodes, the function returns -1. 

# This approach ensures an efficient O(n) time complexity with a straightforward queue-based BFS and a hash set for constant-time lookups.
```
