```python
import hashlib

def hash_to_range(input_string: str) -> int:
     hash_object = hashlib.sha256(input_string.encode())
     hash_int = int(hash_object.hexdigest(), 16)
     return (hash_int % 3) + 1
input_string = "nithursan"
result = hash_to_range(input_string)
print(result)
```


-   Code the solution to your assigned problem in Python (code chunk). Try to find the best time and space complexity solution!


```python
from typing import List

def missing_num(nums: List[int]) -> List[int]:

    # Check for empty list
    if not nums:
        return -1
    
    n = max(nums)  # Find maximum number in the list

    # Obtain full range of numbers
    full_range = set(range(n + 1))
    
    # Convert list to set (also removes duplicates)
    numbers = set(nums)  
    
    # Find missing numbers in sorted order
    missing = sorted(full_range - numbers)  
    
    # Return missing numbers (or -1 if none)
    return missing if missing else -1

lst = [2,7,5]
missing_num(lst)
```

```python
from typing import List

def missing_num(nums: List) -> int:

    # Check for empty list
    if not nums:
        return -1
    
    # Sort the list of numbers
    nums.sort()

    # Find the range of values with max number (upper limit)
    end = nums[-1]
    
    # Generate the full range of values
    full_range = set(range(0, end + 1))
    
    # Find the missing values by getting the difference between the 2 sets
    missing_values = full_range - set(nums)

     # Return -1 if nothing is missing, otherwise return a sorted list
    return sorted(missing_values) if missing_values else -1

lst = [1,2,10,3]
missing_num(lst)
```
